```python
#Importing Pandas
import pandas as pd

#import Numpy
import numpy as np
```

```python
# Loading our dataset

cars = pd.read_csv('http://bit.ly/autolib_dataset', encoding='latin1')
cars.head(10)
```

```python
# Accesssing the information about our dataset
cars.info()
```

```python
# Displaying all rows

pd.set_option('display.max_rows', cars.shape[0])
cars.head(10)
```

```python
#Reading the total null values

cars.isnull().sum()
```

```python
# Dropping the columns where there are missing values

cars.drop(["Displayed comment", "Scheduled at"], axis = 1, inplace = True)

# Displaying our dataset again
cars.head(5)
```

```python
# Checking for duplicates

cars_dupl = cars.drop_duplicates()
cars_dupl.head(10)
```

```python
# Standardization of the columns
#Stripping White Spaces, placing small caps.
cars.columns = cars.columns.str.strip().str.lower().str.replace(' ', '_').str.replace('(', '').str.replace(')', '')
cars.head(10)
```

```python
# Identify the most popular hour of the day 
#for picking up a shared electric car (Bluecar) 
#in the city of Paris over the month of April 2018.

# cars_car = cars.bluecar_counter
# cars_car.head(20)

cars_hour = cars.hour[(cars['city'] == 'Paris') & (cars['month'] == 4) & (cars['year'] == 2018) & (cars['rental_status'] == 'operational') & (cars.bluecar_counter)].mode()
cars_hour
```

The most popular hour for picking up a shared electric car (Bluecar) is 12 noon

```python
# What is the most popular hour for returning cars?

cars_pop_hour = cars.hour[cars['slots']].mode()
cars_pop_hour
```

The Most popular hour of the day is 1100 hrs

```python
# What station is the most popular?
# Overall?

cars_station_pop = cars['station_type'].mode()
print(cars_station_pop)

print("*********************************")

# At the most popular picking hour?

cars_station_pop_hr = cars.station_type[cars['hour']].mode()
cars_station_pop_hr
```

The Most Poplar station is 'station'
While the most popular station at the most popular picking hour is still 'Station'

```python
#What postal code is the most popular for picking up Blue cars? Does the most popular station belong to that postal code?
#Overall?

cars_postal = cars.postal_code[cars['bluecar_counter']].mode()
print(cars_postal)

cars_postal_pop = cars[(cars['postal_code'] == 75015) & (cars['station_type'] == 'station')]
cars_postal_pop.head(10)
```

We can say Here that the postal code is the most popular for picking up Blue cars is 75015 and it uses the same station_type which is station

```python
# At the most popular picking hour?

cars_postal_pop = cars.hour[(cars['postal_code'] == 75015) & (cars['station_type'] == 'station')].mode()
cars_postal_pop
```

No it is not since the most popular picking hour

```python
# Do the results change if you consider Utilib and Utilib 1.4 instead of Blue cars? 

cars_utilib = cars.postal_code[cars['utilib_counter']].mode()
print(cars_utilib)

cars_utilib1 = cars.postal_code[cars['utilib_1.4_counter']].mode()
print(cars_utilib1)
```

Hence the results does not change
